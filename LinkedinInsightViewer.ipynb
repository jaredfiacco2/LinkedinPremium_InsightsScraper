```python
import pandas as pd
import matplotlib.pyplot as plt

# Color Blind Colors - Tableau Pallette
cb_darkblue = '#1170aa'
cb_medblue = '#5fa2ce'
cb_lightblue = '#a3cce9'
cb_maxgrey = '#57606c'
cb_darkgrey = '#7b848f'
cb_medgrey = '#a3acb9'
cb_lightgrey = '#c8d0d9'
cb_tan = '#ffbc79'
cb_orange = '#fc7d0b'
cb_red = '#c85200'
```

```python
df = pd.read_csv("data/companies_list.csv")
#print(len(.index))
df
```

```python
efg = pd.read_pickle("data/emp_func_growth.pkl")
#print(len(efg.index))
efg
#efg['company_name'].unique()
```

```python
ehc = pd.read_pickle("data/emp_head_count.pkl")
#print(len(efg.index))
ehc
ehc['Company Type'] = ehc['sector'].apply(lambda x: 'Linkedin Top 50' if x[:1] == 'L' else 'Healthcare')
ehc_Jan_22_h = ehc.query('`date` == "January 2022"').query('`Company Type` == "Healthcare"').query('`company_name` != "Pfizer"').query('`company_name` != "Walgreens"').groupby(['company_name']).sum()
ehc_Jan_22_n = ehc.query('`date` == "January 2022"').query('`Company Type` != "Healthcare"').groupby(['company_name']).sum()
ehc
```

```python
# Total Employee Head Count Bar Chart - 2021 - Healthcare
plt.figure(figsize=(10, 15))
ehc_cn = ehc_Jan_22_h.index
ec = ehc_Jan_22_h['emp']
nh_1 = plt.barh(ehc_cn, ec, color=cb_darkblue, edgecolor='black')
plt.legend([nh_1], ['Healthcare Companies'], title="Type", loc="upper right")
plt.title("Healthcare's Total Employee Head Count Per Company - Jan 2022")
plt.xlabel('Total Employee Count')
plt.ylabel('Companies')
plt.show()
```

```python
# Total Employee Head Count Bar Chart - 2021 - Linkedin's Top 50
plt.figure(figsize=(10, 15))
ehc_cn = ehc_Jan_22_n.index
ec = ehc_Jan_22_n['emp']/1000
nh_1 = plt.barh(ehc_cn, ec, color=cb_lightblue, edgecolor='black')
plt.legend([nh_1], ["Linkedin's Top 50 Companies"], title="Type", loc="upper right")
plt.title("Linkedin's Top 50 Companies - Total Employee Head Count Per Company - Jan 2022")
plt.xlabel('Total Employee Count (Thousands)')
plt.ylabel('Companies')
plt.show()
```

```python
# New Hires Data Frame
nh = pd.read_pickle("data/new_hires.pkl")
nh['Company Type'] = nh['sector'].apply(lambda x: 'Linkedin Top 50' if x[:1] == 'L' else 'Healthcare')
nh_2021_h = nh.query('`year` == "2021"').query('`Company Type` == "Healthcare"').query('`company_name` != "Pfizer"').query('`company_name` != "Walgreens"').groupby(['company_name']).sum()
nh_2021_n = nh.query('`year` == "2021"').query('`Company Type` != "Healthcare"').groupby(['company_name']).sum()
nh
```

```python
# Plot Horizontal Bar Chart - Healthcare - Total New Hires 2021
plt.figure(figsize=(10, 15))
nh_cn = nh_2021_h.index
tnh = nh_2021_h['senior_employees'] + nh_2021_h['other_employees']
nh_1 = plt.barh(nh_cn, tnh, color=cb_darkblue, edgecolor='black')
plt.legend([nh_1], ['Total New Hires'], title="Type", loc="upper right")
plt.title("Healthcare's New Hires Per Company - Total for 2021")
plt.xlabel('New Hire Count')
plt.ylabel('Companies')
plt.show()
```

```python
# Plot Horizontal Bar Chart - Linkedin's Top 50 - Total New Hires 2021
plt.figure(figsize=(10, 15))
nh_cn = nh_2021_n.index
tnh = nh_2021_n['senior_employees'] + nh_2021_n['other_employees']
nh_1 = plt.barh(nh_cn, tnh, color=cb_lightblue, edgecolor='black')
plt.legend([nh_1], ['Total New Hires'], title="Type", loc="upper right")
plt.title("Linkedin's Top 50 Companies New Hires Per Company - Total for 2021")
plt.xlabel('New Hire Count')
plt.ylabel('Companies')
plt.show()
```

```python
# Total Job Openings By Function Data Frame
tjo = pd.read_pickle("data/total_job_openings.pkl")
tjo['employee_count'] = tjo['employee_count'].astype(float)
tjo['Company Type'] = tjo['sector'].apply(lambda x: 'Linkedin Top 50' if x[:1] == 'L' else 'Healthcare')
tjo
```

```python
# Make Arrays for Companies, and Functions
companies = tjo['company_name'].unique()
lcn = tjo['linkedin_company_name'].unique()
bd,e, it, hs, r, qa, a, ppm, o, pm = [], [], [], [], [], [], [], [], [], []
for c in lcn:
    if c+'_Business Development' in tjo.index:
        tjo_bd = tjo.loc[c+'_Business Development']
        bd.append(tjo_bd['employee_count'])
    else:
        bd.append(0)
    
    if c+'_Engineering' in tjo.index:
        tjo_e = tjo.loc[c+'_Engineering']
        e.append(tjo_e['employee_count'])
        
    else:
        e.append(0) 
    if c+'_Information Technology' in tjo.index:
        tjo_it = tjo.loc[c+'_Information Technology']
        it.append(tjo_it['employee_count'])
    else:
        it.append(0) 
    if c+'_Healthcare Services' in tjo.index:
        tjo_hs = tjo.loc[c+'_Healthcare Services']
        hs.append(tjo_hs['employee_count'])
    else:
        hs.append(0) 
    if c+'_Research' in tjo.index:
        tjo_r = tjo.loc[c+'_Research']
        r.append(tjo_r['employee_count'])
    else:
        r.append(0) 
    if c+'_Quality Assurance' in tjo.index:
        tjo_qa = tjo.loc[c+'_Quality Assurance']
        qa.append(tjo_qa['employee_count'])
    else:
        qa.append(0) 
    if c+'_Administrative' in tjo.index:
        tjo_a = tjo.loc[c+'_Administrative']
        a.append(tjo_a['employee_count'])
    else:
        a.append(0)
    if c+'_Program and Project Management' in tjo.index:
        tjo_ppm = tjo.loc[c+'_Program and Project Management']
        ppm.append(tjo_ppm['employee_count'])
    else:
        ppm.append(0) 
    if c+'_Operations' in tjo.index:
        tjo_o = tjo.loc[c+'_Operations']
        o.append(tjo_o['employee_count'])
    else:
        o.append(0) 
    if c+'_Product Management' in tjo.index:
        tjo_pm = tjo.loc[c+'_Product Management']
        pm.append(tjo_pm['employee_count'])
    else:
        pm.append(0) 
# Plot Stacked Bar Chart
plt.figure(figsize=(10, 10))
b1 = plt.barh(companies, bd, color=cb_tan)
b2 = plt.barh(companies, e, left=bd, color=cb_orange)
b3 = plt.barh(companies, it, left=e, color=cb_red)
b4 = plt.barh(companies, hs, left=it, color=cb_darkblue)
b5 = plt.barh(companies, r, left=hs, color=cb_medblue)
b6 = plt.barh(companies, qa, left=r, color=cb_lightblue)
b7 = plt.barh(companies, a, left=qa, color=cb_lightgrey)
b8 = plt.barh(companies, ppm, left=a, color=cb_medgrey)
b9 = plt.barh(companies, o, left=ppm, color=cb_darkgrey)
b10 = plt.barh(companies, pm, left=o, color=cb_maxgrey)
plt.legend([b1, b2, b3, b4, b5, b6, b7, b8, b9, b10], ['Business Development', 'Engineering', 'Information Technology', 'Healthcare Services', 'Research', 'Quality Assurance', 'Administrative', 'Program and Project Management', 'Operations', 'Product Management'], title="Functions", loc="upper right")
plt.title('Total Job Openings By Functions (Stacked) Per Company ')
plt.xlabel('Total Job Openings')
plt.ylabel('Companies')
plt.show()
```

```python
# Median Tenure Data Frame
mt = pd.read_pickle("data/med_tenure.pkl")
mt['median_tenure_yr'] = mt['median_tenure'].str.replace(' years', '')
mt['median_tenure_yr'] = mt['median_tenure_yr'].astype(float)
mt['color'] = mt['sector'].apply(lambda x: cb_lightblue if x[:1] == 'L' else cb_darkblue)
mt['Company Type'] = mt['sector'].apply(lambda x: 'Linkedin Top 50' if x[:1] == 'L' else 'Healthcare')
mt
```

```python
###################################
### Median Tenure Distributions ###
###################################
# Median Tenure Distribution - Top 50
mt_T = mt.query('`sector` != "Healthcare"')
plt.hist(mt_T['median_tenure_yr'], bins=15, edgecolor='black', color=cb_lightblue)
plt.title('Median Tenure Histogram - Top 50 Companies (According to Linkedin)')
plt.xlabel('Median Tenure (Years)')
plt.ylabel('Bin Frequency')
plt.xlim([0, 10])
plt.show()
```

```python
# Median Tenure Distribution - Healthcare
mt_H = mt.query('`sector` == "Healthcare"')
plt.hist(mt_H['median_tenure_yr'], bins=15, edgecolor='black', color=cb_darkblue)
plt.title('Median Tenure Histogram - Healthcare Companies in the Tri-State Area')
plt.xlabel('Median Tenure (Years)')
plt.ylabel('Bin Frequency')
plt.xlim([0, 10])
plt.show()
```

```python
# Median Tenure Distribution - All
plt.hist(mt['median_tenure_yr'], bins=15, edgecolor='black', color=cb_darkgrey)
plt.title('Median Tenure Histogram - All')
plt.xlabel('Median Tenure (Years)')
plt.ylabel('Bin Frequency')
plt.xlim([0, 10])
plt.show()
```

```python
# Median Tenure - Stacked Histogram
mt_s = mt 
mt.pivot(columns='Company Type').median_tenure_yr.plot(kind = 'hist', stacked=True, bins=15, color = [cb_lightblue, cb_darkblue], edgecolor = 'black', title='Stacked Median Tenure Histogram - Top 50 Compaies & Healthcare Companies in the Tri-State Area', ylabel='Bin Frequency', xlabel='Median Tenure (Years)')
```

```python
# Median Tenure - Sorted Horizontal Bar Graph - Healthcare
mt_H = mt_H.sort_values('median_tenure_yr')
plt.figure(figsize=(10, 8))
h = plt.barh(mt_H['company_name'],mt_H['median_tenure_yr'], edgecolor='black', color=mt_H['color'])
plt.title('Median Tenure Histogram - Healthcare Companies in the Tri-State Area')
plt.xlabel('Median Tenure (Years)')
plt.ylabel('Company')
plt.xlim([0, 10])
plt.legend(h, ('Healthcare Companies',), loc='lower right')
plt.show()
```
